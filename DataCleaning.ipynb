```python
import pandas as pd
```

```python
df=pd.read_csv("GlassdoorSalary.csv")
df.head()
```

```python
#salary parsing
#Company name Text only
#state field
#age of company
#parsing of job description
```

### Salary Parsing

```python
df["hourly"]=df["Salary Estimate"].apply(lambda x:1 if "per hour" in x.lower() else 0)
```

```python
df["employer_provided"]=df["Salary Estimate"].apply(lambda x:1 if "employer provided salary" in x.lower() else 0)
```

```python
df=df[df["Salary Estimate"]!="-1"]
```

```python
salary=df["Salary Estimate"].apply(lambda x:x.split("(")[0])
```

```python
salary
```

```python
minus_kd=salary.apply(lambda x:x.replace("K","").replace("$",""))
```

```python
minus_kd
```

```python
df.head()
```

```python
df.head()
```

```python
min_hr=minus_kd.apply(lambda x:x.lower().replace("per hour","").replace("employer provided salary","").replace(":",""))
```

```python
df["min_salary"]=min_hr.apply(lambda x:int(x.split("-")[0]))
df["max_salary"]=min_hr.apply(lambda x:int(x.split("-")[1]))
```

```python
df.head()
```

```python
df["avg_salary"]=(df.min_salary+df.max_salary)/2
```

### Company Name

```python
df["Company_Text"]=df.apply(lambda x:x["Company Name"] if x["Rating"]<0 else x["Company Name"][:-3],axis=1)
```

```python
df.head()
```

```python
df["Company_Text"]=df["Company_Text"].apply(lambda x:x.replace("\r\n",""))
```

```python
df.head()
```

### State Field

```python
df["job_state"]=df["Location"].apply(lambda x:x.split(",")[1])
```

```python
df.head()
```

```python
df.job_state.value_counts()
```

```python
df["same_state"]=df.apply(lambda x: 1 if x.Location==x.Headquarters else 0,axis=1)
```

### Age Of Company

```python
df["age"]=df.Founded.apply(lambda x:x if x<1 else 2022-x)
```

```python
df.head()
```

### Parsing of job description

```python
df["python_yn"]=df["Job Description"].apply(lambda x:1 if "python" in x.lower() else 0)
df.python_yn.value_counts()
```

```python
df["r_yn"]=df["Job Description"].apply(lambda x:1 if "r studio" in x.lower() or "r-studio" in x.lower() else 0)
df["r_yn"].value_counts()
```

```python
df["spark_yn"]=df["Job Description"].apply(lambda x:1 if "spark" in x.lower() else 0)
df["spark_yn"].value_counts()
```

```python
df["aws_yn"]=df["Job Description"].apply(lambda x:1 if "aws" in x.lower() else 0)
df["aws_yn"].value_counts()
```

```python
df["excel_yn"]=df["Job Description"].apply(lambda x:1 if "excel" in x.lower() else 0)
df["excel_yn"].value_counts()
```
